```python
#This is to fill the remaining empty cases under the various columns with their respective placeholders
tree_census_subset['problems'].fillna('None', inplace=True)
tree_census_subset['health'].fillna('Good', inplace=True)
tree_census_subset['spc_latin'].fillna('No Observation', inplace=True)
tree_census_subset['sidewalk'].fillna('NoDamage', inplace=True)
```

```python
big_trees = tree_census_subset[(tree_census_subset['tree_dbh']>60) | (tree_census_subset['stump_diam']>60)]
big_trees
```

```python
tree_census_subset2 = tree_census_subset[(tree_census_subset['tree_dbh']<=60) | (tree_census_subset['stump_diam']<=60)]
tree_census_subset2
```

```python
tree_census_subset2_alive = tree_census_subset2[tree_census_subset2['status']=='Alive']
tree_census_subset2_dead_or_stump = tree_census_subset2[(tree_census_subset2['status']=='Dead') | (tree_census_subset2['status']=='Stump')]
```

```python
tree_census_subset2_alive.groupby('spc_latin').mean()
```

```python
tree_census_subset2_alive.groupby('spc_latin')['tree_dbh'].describe()
```

```python
stats_alive = tree_census_subset2_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive
```

```python
tree_census_subset_alive3 = tree_census_subset2_alive.merge(stats_alive, on='spc_latin', how='left')
tree_census_subset_alive3
```

```python
mask = tree_census_subset_alive3['tree_dbh'] < tree_census_subset_alive3['25%']
tree_census_subset_alive3.loc[mask, 'tree_dbh'] = tree_census_subset_alive3['25%']

mask = tree_census_subset_alive3['tree_dbh'] > tree_census_subset_alive3['75%']
tree_census_subset_alive3.loc[mask, 'tree_dbh'] = tree_census_subset_alive3['75%']
```
